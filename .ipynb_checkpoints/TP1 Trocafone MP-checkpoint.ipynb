```python
import datetime as datetime
import datetime 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
dataTroca = pd.read_csv('events.csv') 
dataTroca.info()
```

```python
#Convierto la fecha de creacion en un datetime64[ns]
dataTroca['timestamp'] = pd.to_datetime(dataTroca['timestamp'])

#Se ve que todas las entradas son del 2018
dataTroca.groupby(dataTroca.timestamp.dt.year).count()
```

```python
groupbyPersonas = dataTroca.loc[:,['person', 'event']].groupby('person')
dicc = {}
for clave, data in groupbyPersonas:
    lista = []
    lista = data.drop_duplicates(subset = 'event', keep = 'first')['event'].tolist()
    dicc[clave] = lista
```

```python
#Obtengo una lista de todos los eventos que tiene esa persona
dataTroca['eventos'] = dataTroca['person'].map(lambda x: dicc[x])
#Obtengo los dias de la seman
dataTroca['dayweek'] = dataTroca.timestamp.dt.dayofweek
#Obtengo la fecha
dataTroca['fecha'] = dataTroca.timestamp.dt.date
#Separo la marca
dataTroca['marca'] = dataTroca['model'].map(lambda x: str(x).split()[0])
```

```python
#cantidad de personas que ingresaron 27624.
groupbyCantEventos = dataTroca.groupby('person', as_index = False)['event'].agg({'cantEventos': 'count'})
groupbyCantEventos.sort_values(by = 'cantEventos', ascending = False)
dataTroca = pd.merge(dataTroca, groupbyCantEventos, on = 'person')
```

```python
#Filtro a la gente que no tiene el evento visited_site 
personasConVisitas = dataTroca.loc[dataTroca['event'] == 'visited site']\
                              .drop_duplicates(subset = 'person', keep = 'first')['person'].tolist()
dataNotVisit = dataTroca.loc[dataTroca['person'].isin(personasConVisitas)]
```

```python
data2 = dataNotVisit.loc[dataNotVisit['event'] == 'visited site'].groupby(['person', 'fecha'], as_index = False).agg({'city': 'first', 'region': 'first', 'country': 'first', 'device_type': 'first', 'operating_system_version': 'first', 'channel': 'first', 'new_vs_returning': 'first'})
```

```python
data3 = pd.merge(dataNotVisit, data2, on = ['person', 'fecha'])
data3.drop(['region_x', 'device_type_x', 'operating_system_version_x', 'country_x', 'city_x', 'channel_x', 'new_vs_returning_x'], axis = 1, inplace = True)
data3.rename(columns = {'region_y': 'region', 'device_type_y': 'device_type', 'operating_system_version_y': 'SO', 'city_y': 'city', 'country_y': 'country', 'channel_y': 'channel', 'new_vs_returning_y': 'new_vs_returning'}, inplace =True)
```

```python
def obtenerSistema(so):
    sistema = so.split()
    if len(sistema) > 1:
        otro = sistema[1].split('.')[0]
        return sistema[0] + ' ' + otro
    else:
        return sistema[0]
```

```python
data3['SO'] = data3['SO'].map(obtenerSistema)
```

```python
#Agrupo los eventos para ver cuan efectiva fue la campaÃ±ana de publicidad y por lo visto fue bastante buena siendo
#el 4 evento mas repetido.
groupByEvent = data3.loc[(data3['event'] == 'visited site') | (data3['event'] == 'ad campaign hit') | (data3['event'] == 'search engine hit'),:].groupby('event')['person'].agg({'cantPersonas': 'count'})
g = groupByEvent.sort_values(by = 'cantPersonas', ascending = False).apply(lambda x: x*100 / x.sum()).plot(kind='bar',figsize=(14,4));
g.set_title("Porcentaje de eventos de acceso", fontsize=18)
g.set_xlabel("Evento",fontsize=14)
g.set_ylabel("Cantidad de personas", fontsize=14)
```

```python
#Creo un nuevo Ã­ndice para aplicar mÃ¡s adelante.
mes = {1:'Enero', 2:'Febrero', 3:'Marzo', 4:'Abril', 5:'Mayo', 6:'Junio'}
```

```python
#Quiero ver la cantidad de ventas por mes que tuvo la empresa.
dataTrocaPerMonth = data3[['timestamp', 'event']]
dataTrocaPerMonth['month'] = dataTrocaPerMonth['timestamp'].apply(lambda x: x.month)
g = dataTrocaPerMonth.loc[dataTrocaPerMonth['event'] == 'conversion',['event','month']].groupby('month').count().plot(figsize=(14,4));
g.set_title("Cantidad de ventas por mes", fontsize=18)
g.set_xlabel("Mes",fontsize=14)
g.set_ylabel("Cantidad de ventas", fontsize=14)
```

```python
#Quise mostrar el ratio de ventas que hubo por mes para ver si la empresa habia crecido.
#Por lo visto la empresa estaba en equilibrio hasta el mes de Marzo donde las ventas disminuyeron fuertemente.
#Esto se puede deber al aumento descomunal que tuvo el dolar en estos ultimos meses.
namesgrouped = dataTrocaPerMonth.groupby(['month','event'])['timestamp'].agg({'ratioVentas': 'count'})
ratiospereventpermonth = namesgrouped.groupby(level=[0]).apply(lambda x: x / x.sum())
g = ratiospereventpermonth.filter(like='conversion', axis=0).plot(figsize=(14,4));
g.set_title("Ratio de ventas por mes", fontsize=18)
g.set_xlabel("Mes",fontsize=14)
g.set_ylabel("Ratio de ventas", fontsize=14)
```

```python
#Quiero mostrar la cantidad de eventos que ocurren dependiendo de la hora.
dataTrocaPerHour = data3.loc[:,['timestamp','event']]
dataTrocaPerHour['hour'] = dataTrocaPerMonth['timestamp'].apply(lambda x: x.hour)
dataTrocaPerHour.loc[:,['event','hour']].groupby('hour').agg(np.size).plot(kind='bar',figsize=(14,4));
```

```python
#Casi la mitad de los eventos son realizados desde celulares. Por ahi seria una buena idea, en el caso que no este
#implementada, crear una aplicacion para celular para facilitarle las cosas a los usuarios de Smartphones.
dataPerDevice = data3.loc[:,['event','device_type']].groupby('device_type').agg(np.size)
g = dataPerDevice.apply(lambda x: x*100 / x.sum()).filter(items=['Computer', 'Smartphone'], axis=0).plot.pie(y='event',figsize=(5, 5))
g.set_title("Dispositivos utilizados para navegar", fontsize=18)
g.set_ylabel("Porcentaje de uso", fontsize=14)
```

```python
#Quiero ver la cantidad de personas que habiendose conectado con Android, terminaron comprando un celular con iOS.
dataSOAndroid = data3.loc[(data3.event.str.contains('conversion')) & (data3.SO.str.contains('Android')),:]
dataSOAndroid['changeSO'] = np.where(dataSOAndroid['model'].str.contains('iPhone'), 'yes', 'no')
g = dataSOAndroid.loc[:,['event','changeSO']].groupby('changeSO').agg(np.size).apply(lambda x: x*100 / x.sum()).plot.pie(y='event',figsize=(5, 5))
g.set_title("Cambios de SO de Android a iOS", fontsize=18)
g.set_ylabel("Porcentaje de cambio", fontsize=14)
```

```python
#Quiero ver la cantidad de personas que habiendose conectado con iOS, terminaron comprando un celular con Android.
dataSOiOS = data3.loc[(data3.event.str.contains('conversion')) & (data3.SO.str.contains('iOS')),:]

dataSOiOS['changeSO'] = np.where(dataSOiOS['model'].str.contains('iPhone'), 'no', 'yes')
g = dataSOiOS.loc[:,['event','changeSO']].groupby('changeSO').agg(np.size).apply(lambda x: x*100 / x.sum()).plot.pie(y='event',figsize=(5, 5))
g.set_title("Cambios de SO de iOS a Android", fontsize=18)
g.set_ylabel("Porcentaje de cambio", fontsize=14)
```
